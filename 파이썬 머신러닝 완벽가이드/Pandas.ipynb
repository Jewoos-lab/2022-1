## DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

### 넘파이 ndarray, 리스트 딕셔너리를 DataFrame으로 변환하기

```python
import numpy as np

# 1차원 데이터를 기반으로 DataFrame 생성

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
# 2차원 데이터를 기반으로 DataFrame생성

col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
        [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)  # 2행 3열
print()

df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
print()

df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# 딕셔너리를 DataFrame으로 변환

# key는 문자열 칼럼명으로 매핑, Value는 리스트형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1':[1, 11], 'col2':[2, 22], 'col3':[3, 33]}
df_dict = pd.DataFrame(dict)
print("딕셔너리로 만든 DataFrame:\n",df_dict)  # key는 칼럼명, value는 칼럼 데이터로 매핑됐다.
```

### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

```python
df_dict.values
```

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), '\ndf_dict.values shape:', array3.shape)
print(array3)
```

```python
# DataFrame을 리스트로 변환 
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)
```

```python
# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

## DataFrame의 칼럽 데이터 세트 생성과 수정

```python
# 첫번째행 첫번째열
data_df.iloc[0, 0]

# 행과 열위치에 위치 기반 인덱싱 값을 입력하면 된다.
```

```python
# 마찬가지로 문자열 인덱스를 행위치에 입력해도 오류가 발생한다.
data_df.iloc['one', 0]
```

### DataFrame loc[ ]연산자
- loc[ ]는 명칭 기반으로 데이터를 추출합니다. 따라서 행 위치에는 DataFrame index값을, 그리고 열 위치에는 칼럼 명을 입력해 줍니다.

```python
# 인덱스값이 'one', 행 칼럼명이 'name'
data_df.loc['one', 'Name']
```

```python
data_df_reset.loc[0, 'Name']
```

```python
print('위치 기반 iloc slicing\n', data_df.iloc[0:1, 0], '\n')
print('명칭 기반 loc slicing\n', data_df.loc['one':'two', 'Name'])
```

- iloc[ ]는 위치기반 인덱싱만 가능. 따라서 행과 열 위치 값으로 정수형 값을 지정해 원하는 데이터를 반환한다.
- loc[ ]는 명칭 기반 인덱싱만 가능. 따라서 행 위치에 DataFrame 인덱스가 오며, 열 위치에는 칼럼 명을 지정해 원하는 데이터를 반환한다.
- 명칭 기반 인덱싱에서 슬라이싱을 [시작점:종료점]으로 지정할때 시작점에서 종료점을 포함한 위치에 있는 데이터를 반환한다.

### fillna()로 결손데이터 대체하기

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
# Age칼럼의 NaN값을 평균나이, Embarked 칼럼의 NAN값을 S로 대체해 모든 결손 데이터를 처리하겠다
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].mean())
titanic_df["Embarked"] = titanic_df["Embarked"].fillna("S")
titanic_df.isna().sum()
```

## apply lambda 식으로 데이터 가공

```python
def get_square(a):
  return a**2

print("3의 제곱:", get_square(3))
```

```python
lambda_square = lambda x : x **2
print(lambda_square(3))
```

```python
# 람다식을 사용할때 여러개의 값을 입력인자로 사용해야 할 경우에는 보통 map()함수 결합해서 사용

a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[["Name", "Name_len"]].head(3)
```

```python
# 나이가 15세 미만 chile 그렇지 않으면 adult구분
titanic_df["Child_Adult"] = titanic_df["Age"].apply(lambda x : 'Child' if x <=15 else "Adult")
titanic_df[["Age", "Child_Adult"]].head(8)
```

```python
# 15세 이하면 Child 15~60세 Adult, 61세 이상은 Elderly
titanic_df["Age_cat"] = titanic_df["Age"].apply(lambda x : "Child" if x <= 15 else ("Adult" if x <= 60 else "Elderly"))
titanic_df["Age_cat"].value_counts()
```
